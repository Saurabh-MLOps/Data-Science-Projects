```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [0,1,2,3,4,5]
y = [0,2,4,6,8,10]

#Resize your graph(dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize =(5,3), dpi = 150 )


#plt.plot(x,y, label = '2x', color = 'red', linewidth = 2, marker = '.',markersize = 10,linestyle = '--', markeredgecolor = 'green')


#Use shorthand notation
# fmt = '[color][marker][Line]'
plt.plot (x,y, 'b^--',label = '2x')

## Line number two

# Select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

#Plot part of the graph as line
plt.plot(x2[:5], x2[:5]**2,'r', label = 'x^2')

#Plot remainder of graph 
plt.plot(x2[4:], x2[4:]**2, 'r--')

#Add a title (specify font parameters with fontdict)
plt.title ('Our First Graph!', fontdict ={'fontname': 'Comic Sans MS', 'fontsize': 20})   #Resize the graph

# X and Y labels  
plt.xlabel('X axis', fontdict= {'fontsize': 20 })     #Increasing fontsize of label x-axis
plt.ylabel('Y axis')

plt.legend()     # To add legend in our graph i.e '2x'


#To save your graph(dpi 300 is good when saving so graph has high resolution)
plt.savefig('graph1.png', dpi = 300)

plt.show()

```

### Bar Chart

```python
labels = ['A','B','C']
values = [1,4,3]

bars = plt.bar(labels, values)

bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch('*')

plt.figure(figsize = (6,4))

plt.show()
```
